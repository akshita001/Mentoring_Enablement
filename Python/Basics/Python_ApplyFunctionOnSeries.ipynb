```python
# Transformations
# Categorize those cars that are more than 4 cylinders

cars_data['MoreThan4_Cylinders'] = cars_data['cyl']\
    .map({4: 'No',
         6:'Yes',
         8:'Yes'})
```

```python
# Use a named function

def drop_comma(x):
    return str(x).replace(',','')
```

```python
cars_data['mpg'].head()
```

```python
cars_data['mpg'] = cars_data['mpg'].apply(drop_comma)
```

```python
cars_data['mpg'].head()
```

```python
cars_data['mpg'] = pd.to_numeric(cars_data['mpg'])
```

```python
# One hot encoding or another transformed column added based on some logic

def category(cyl):
    
    cyl = int(cyl)
    
    if cyl <= 4:
        return 'Low Cylinders'
    elif cyl >4 and cyl <= 8:
        return 'Medium Cylinders'
    elif cyl >= 8:
        return 'High Cylinders'
    else:
        return 'NA'
```

```python
cars_data.cyl.apply(category)
```
