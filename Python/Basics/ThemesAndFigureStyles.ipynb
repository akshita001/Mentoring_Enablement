#### Set default aesthetic parameters
The sns.set() function allows us to set the parameters for the appearance of our plot. By calling it without explicitly mentioning any parameters, we end up with the default values. These include:
* context = 'notebook'
* style = 'darkgrid'
* palette = 'deep'
* font = 'sans-sherif'

```python
sns.set()
```

```python
fig,axis = plt.subplots(figsize = (10,5))

sns.distplot(pollution_data["NO2.AQI"])

plt.show()
```

*As soon as we set the plot to its default and plotted the graph again, we notice the change in pattern. Shows that our notebook was not in the default mode earlier*

Seaborn has five preset styles - darkgrid, whitegrid, dark, white and ticks. Let us see each of these one by one.

#### Plain dark background

```python
sns.set_style("dark")

sns.distplot(pollution_data["NO2.AQI"])

plt.show()
```

#### White grid on a dark background. This is the default mode.

```python
sns.set_style("darkgrid")

sns.distplot(pollution_data["NO2.AQI"])

plt.show()
```

#### Passing parameters in a dictionary
The grid style parameters can be set in a dictionary object. Here, we set the background color of the grid by using the axes.facecolor attribute. The value will range from 0 (white) to 1 (black)

```python
sns.set_style("darkgrid" ,
             {"axes.facecolor": "0.8"})

sns.distplot(pollution_data["NO2.AQI"])

plt.show()
```

By changing the axes.facecolor we can change the contrast of the background.By reducing the facecolor towards 0 the background become more darker,color almost become negative. 

#### Plain white background

```python
sns.set_style("white")

sns.distplot(pollution_data["NO2.AQI"])

plt.show()
```

#### Whitegrid - white background, with grids on the plot

```python
sns.set_style("whitegrid")

sns.distplot(pollution_data["NO2.AQI"])

plt.show()
```

#### Add ticks along the axes

```python
sns.set_style("ticks")

sns.distplot(pollution_data["NO2.AQI"])

plt.show()
```

** For all the above settings, the graph style is set for all plots that follow. What if we want to use a different style for just a single plot, and then fall back to the original style?**

```python
with sns.axes_style("whitegrid"):

    sns.distplot(pollution_data["NO2.AQI"])
    
plt.show()
```

#### The despine() feature

```python
sns.distplot(pollution_data["NO2.AQI"])

sns.despine()

sns.set_style("whitegrid")

plt.show()
```

Following from what we said above, as the last axes style was set to ticks, we see that the above plot has retained the ticks style, but we see that there is no box surrounding the plot. The right and top lines have been despined.

#### Apply the despining to all edges
We end up with a spineless plot

```python
sns.distplot(pollution_data["NO2.AQI"])

sns.despine(left=True, 
            right=True, 
            bottom=True, 
            top=True)

plt.show()
```

#### Apply an offset to the axes or edges

```python
sns.distplot(pollution_data["NO2.AQI"])

sns.despine(offset=25,
            trim=False)

plt.show()
```

There is an additional 'trim' parameter that can be used while making our plots. This is what Seaborn has to say about the feature:
"Some plots benefit from offsetting the spines away from the data, which can also be done when calling despine(). When the ticks donâ€™t cover the whole range of the axis, the trim parameter will limit the range of the surviving spines."
